```python
# Dropando colunas que não serão utilizadas para análise
df.drop(columns=['excess_mortality', 'excess_mortality_cumulative', 'excess_mortality_cumulative_absolute', 'total_boosters_per_hundred', 'total_boosters', 'tests_units', 
                 'hospital_beds_per_thousand','handwashing_facilities', 'male_smokers', 'female_smokers','extreme_poverty', 'gdp_per_capita', 'aged_70_older',
                 'aged_65_older', 'median_age', 'population_density','stringency_index','new_vaccinations_smoothed', 'tests_units','tests_per_case', 'positive_rate',
                 'new_tests_smoothed_per_thousand','new_tests_smoothed', 'new_tests_per_thousand', 'total_tests', 'new_tests', 'weekly_hosp_admissions', 'weekly_icu_admissions_per_million',
                 'weekly_icu_admissions', 'hosp_patients_per_million', 'hosp_patients', 'icu_patients_per_million', 'icu_patients', 'reproduction_rate', 'new_deaths_smoothed_per_million',
                 'new_deaths_per_million', 'new_cases_smoothed_per_million', 'new_cases_per_million', 'new_deaths_smoothed','new_deaths', 'total_deaths', 'new_cases_smoothed',
                 'new_cases', 'iso_code'], inplace=True)
```

```python
# Dropando mais algumas colunas
df.drop(columns=['human_development_index', 'total_vaccinations_per_hundred','new_vaccinations', 'total_vaccinations', 'total_tests_per_thousand', 'weekly_hosp_admissions_per_million', 'diabetes_prevalence',
                 'cardiovasc_death_rate'], inplace=True)
```

```python
# Checando número de colunas e linhas
df.shape
```

```python
# Checando os valores nulos
df.isnull().sum()
```

```python
# Checando a tipagem dos meus dados
df.dtypes
```

```python
# Transformando a data em data e passando para o pandas o formato em que se encontram
df["date"] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
# Checando a alteração
df.dtypes
```

```python
# Criando colunas month, day e year
df["month"], df['day'], df['year'] = (df['date'].dt.month, df['date'].dt.day, df['date'].dt.year)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Pegando os meses cujo valor são diferentes de setembro
indexdrop = df[df['month'] != 9 ].index
# Deletando essas linhas
df.drop(indexdrop, inplace = True)
df.head()
```

```python
# Pegando os anos cujo valor são diferentes de 2021
indexdrop = df[df['year'] != 2021 ].index
# Deletando essas linhas
df.drop(indexdrop, inplace = True)
df.tail()
```

```python
# Pegando os dias menores que 23 (último dia deste dataset é 23/09/2021)
indexdrop = df[df['day'] < 23 ].index
# Deletando essas linhas
df.drop(indexdrop, inplace = True)
df.head()
```

```python
# Pessoas totalmente vacinadas por continente
df.groupby(df["continent"])["people_fully_vaccinated"].sum()
```

```python
# Pessoas totalmente vacinadas no Brazil
df.loc[(df["location"] == "Brazil")]
```

```python
# Criando gráficos 
df.groupby(df["continent"])["people_fully_vaccinated"].sum().plot.pie()
```

```python
# Nº de pessoas vacinadas por centena
df.groupby(df['continent'])['people_fully_vaccinated_per_hundred'].mean().plot.barh()
```

```python
# Total de locais por continente neste dataset
df["continent"].value_counts()
```

```python
# Adicionando um título e alterando nome dos eixos
import matplotlib.pyplot as plt
df["continent"].value_counts().plot.barh(title = "Total de paises neste dataset", color = "green")
plt.xlabel("Paises neste dataset");
plt.ylabel("Continente");
```

```python
# Alterando o estilo
plt.style.use('ggplot')
```

```python
# Mortes por milhao dividido por continentes + salvando o gráfico como imagem
df.groupby(df["continent"])['total_deaths_per_million'].sum().plot.bar(title = "Mortes por milhao dividido por continentes", color = "red")
plt.xlabel("Continente")
plt.ylabel("Mortes por milhao");
plt.savefig("Gráfico mortes por milhão.png")
```
